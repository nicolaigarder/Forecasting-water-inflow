```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load cleaned weather data
file_path = "/Users/nicolaigarderhansen/Desktop/Bachelorprojekt/Forecasting-water-inflow/data_cleaned_and_interpolated.csv"
df = pd.read_csv(file_path, header=0, parse_dates=["time"], index_col="time")
df.head()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt

# Define hyperparameters
TIME_STEPS = 30   # Past 30 days as input
FUTURE_DAYS_LIST = [1, 5, 14]  # Predict 1, 5, and 14 days ahead
BATCH_SIZE = 8
LEARNING_RATE = 0.0001
EPOCHS = 50  
QUANTILE = 0.925

# Quantile Loss Function
def quantile_loss(q, y_true, y_pred):
    e = y_true - y_pred
    return K.mean(K.maximum(q * e, (q - 1) * e))

# Preprocess Data for Multi-Step Prediction
def preprocess_data(df, time_steps, future_days):
    df = df.dropna()  # Remove missing values before processing
    
    scaler_X = RobustScaler()
    scaler_y = RobustScaler()

    X_scaled = scaler_X.fit_transform(df.drop(columns=["KarupBro"]).select_dtypes(include=[np.number]))
    y_scaled = scaler_y.fit_transform(df["KarupBro"].values.reshape(-1, 1))
    
    X_series, y_series = [], []
    for i in range(time_steps, len(X_scaled) - future_days + 1):
        X_series.append(X_scaled[i - time_steps:i])
        y_series.append(y_scaled[i:i + future_days].flatten())  # Collect future_days ahead
    
    return np.array(X_series), np.array(y_series), scaler_X, scaler_y

# Build LSTM Model for Multi-Step Forecasting
def build_model(time_steps, features, future_days, learning_rate, quantile=QUANTILE):
    model = Sequential([
        LSTM(200, return_sequences=True, input_shape=(time_steps, features), kernel_regularizer=l2(0.001)),
        Dropout(0.3),
        LSTM(100, return_sequences=False, kernel_regularizer=l2(0.001)),
        Dropout(0.3),
        Dense(50, activation="relu", kernel_regularizer=l2(0.001)),
        Dense(future_days)  # Output future_days ahead
    ])
    
    model.compile(optimizer=Adam(learning_rate=learning_rate), 
                  loss=lambda y_true, y_pred: quantile_loss(quantile, y_true, y_pred))
    return model

# Evaluate RMSE
def calculate_rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Store results
rmse_results = {}

for future_days in FUTURE_DAYS_LIST:
    print(f"Training model for {future_days}-day ahead forecasting...")
    
    # Prepare dataset
    X_series, y_series, scaler_X, scaler_y = preprocess_data(df, TIME_STEPS, future_days)
    
    # Split into train, validation, and test
    train_size = int(0.6 * len(X_series))
    val_size = int(0.2 * len(X_series))
    test_size = len(X_series) - train_size - val_size

    X_train, y_train = X_series[:train_size], y_series[:train_size]
    X_val, y_val = X_series[train_size:train_size+val_size], y_series[train_size:train_size+val_size]
    X_test, y_test = X_series[train_size+val_size:], y_series[train_size+val_size:]
    
    # Build and train model
    model = build_model(TIME_STEPS, X_series.shape[2], future_days, LEARNING_RATE)
    
    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE, 
        verbose=1
    )
    
    # Predict on test data
    y_pred = model.predict(X_test)
    
    # Handle NaNs in predictions before inverse transformation
    y_pred = np.nan_to_num(y_pred, nan=0.0)
    y_test = np.nan_to_num(y_test, nan=0.0)
    
    # Inverse transform predictions and actual values
    y_test_inverse = scaler_y.inverse_transform(y_test)
    y_pred_inverse = scaler_y.inverse_transform(y_pred)
    
    # Ensure no NaNs in inverse-transformed data
    y_test_inverse = np.nan_to_num(y_test_inverse, nan=0.0)
    y_pred_inverse = np.nan_to_num(y_pred_inverse, nan=0.0)
    
    # Calculate RMSE for each time step
    rmse_per_step = [calculate_rmse(y_test_inverse[:, i], y_pred_inverse[:, i]) for i in range(future_days)]
    mean_rmse = np.mean(rmse_per_step)
    rmse_results[future_days] = mean_rmse
    print(f"Mean RMSE for {future_days}-day ahead: {mean_rmse:.4f}")
    
# Baseline Persistence Model (Last Observed Value is Future Prediction)
baseline_rmse_results = {}
for future_days in FUTURE_DAYS_LIST:
    rmse_per_step_baseline = [calculate_rmse(y_test_inverse[:, i], y_test_inverse[:, 0]) for i in range(future_days)]
    mean_rmse_baseline = np.mean(rmse_per_step_baseline)
    baseline_rmse_results[future_days] = mean_rmse_baseline
    print(f"Baseline RMSE for {future_days}-day ahead: {mean_rmse_baseline:.4f}")

# Plot RMSE Comparison
plt.figure(figsize=(8, 5))
plt.plot(FUTURE_DAYS_LIST, [rmse_results[k] for k in FUTURE_DAYS_LIST], marker='o', label='LSTM Model')
plt.plot(FUTURE_DAYS_LIST, [baseline_rmse_results[k] for k in FUTURE_DAYS_LIST], marker='s', linestyle='dashed', label='Persistence Baseline')
plt.xlabel("Forecast Horizon (Days)")
plt.ylabel("RMSE")
plt.title("RMSE Comparison: LSTM vs Persistence Baseline")
plt.legend()
plt.grid()
plt.show()
```
